**Q1  Write a Python program to convert kilometers to miles?**

**Ans**


```python
def kilometr_to_miles():
    kilomtr = float(input('Enter the distance in Kilometers\n'))
    miles = float(kilomtr*0.621371)
    print('The Distance in miles is :',miles)
```

```python
kilometr_to_miles()
```

**Q2.  Write a Python program to convert Celsius to Fahrenheit?**


**Ans**


```python
def celsius_to_fahrenheit():
    celsius = float(input('Enter the Temperature in Celsius\n'))
    fahrenheit = float(celsius*9/5+32)
    print('The Temperature in Fahrenheit is :',fahrenheit)
```

```python
celsius_to_fahrenheit()
```

**Q4 Write a Python program to solve quadratic equation?**

**Ans**

Image courtesy https://www.cuemath.com/algebra/quadratic-equations/

```python
import cmath
def quadratic():
    a = float(input('Enter the coefficient of X Square'))
    b = float(input('Enter the coefficient of x'))
    c = float(input('Enter the Constant'))
    discriminant = (b**2)-(4*a*c)
    sol1 = (-b-cmath.sqrt(discriminant))/(2*a)  
    sol2 = (-b+cmath.sqrt(discriminant))/(2*a)  
    print('The solution are {0} and {1}'.format(sol1,sol2)) 
    
```

```python
quadratic()
```

**Q5.  Write a Python program to swap two variables without temp variable?**

**Ans**

```python
def swap(a,b):
    print('The value of a and b before swap are:',a,' and ',b)
    a = a+b
    b = a-b 
    a = a-b
    print('The value of a and b after swap are:',a,' and ',b)
```

```python
swap(34,53)
```
